# Carregar bibliotecas

```python
import pandas as pd
```

# Ler arquivo da fonte de dados

```python
url = 'https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv'
df = pd.read_csv(url)
```

# 1. Preparar o modelo

### 1.1 Separa o df em labels (x) e predict (y)

```python
df.head()
```

```python
x  = df[['home', 'how_it_works', 'contact']]
y = df['bought']
```

### 1.1 Separa o df em treino e teste

```python
treino_x = x[:75]
treino_y = y[:75]

teste_x = x[75:]
teste_y = y[75:]

print('Treino com {} elementos \nTeste com {} elementos'.format(len(treino_x), len(teste_x)))
```

### 1.2 Treinar o modelo

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
```

### 1.3 Fazer previsões

```python
previsoes = modelo.predict(teste_x)
previsoes
```

### 1.4 Mensurar a acurácia do modelo

```python
acuracia = accuracy_score(teste_y, previsoes) * 100
print('A acuracia é de {:.2f}%'.format(acuracia))
```
