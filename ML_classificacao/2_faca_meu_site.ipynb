```python
import pandas as pd
```

```python
uri = 'https://gist.githubusercontent.com/guilhermesilveira/1b7d5475863c15f484ac495bd70975cf/raw/16aff7a0aee67e7c100a2a48b676a2d2d142f646/projects.csv'
df = pd.read_csv(uri)
df.head()
```

```python
troca = {0:1, 1:0}
df['finished'] = df['unfinished'].map(troca)

df.head()
```

```python
import seaborn as sns

sns.scatterplot(x = 'expected_hours', y = 'price', data = df)
```

```python
sns.scatterplot(x = 'expected_hours', y = 'price', data = df, hue = 'finished')
```

```python
sns.relplot(x = 'expected_hours', y = 'price', data = df, col = 'finished', hue = 'finished')
```

#### Criando o modelo de ML

```python
#Features
x = df[['expected_hours', 'price']]
#Target
y = df['finished']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
SEED = 20
```

```python
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state = SEED, stratify = y,  test_size = 0.25)

print('Treino com {} elementos \nTeste com {} elementos'.format(len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)

previsoes = modelo.predict(teste_x)
previsoes

acuracia = accuracy_score(teste_y, previsoes) * 100
print('A acuracia é de {:.2f}%'.format(acuracia))
```

#### O teste abaixo é feito com teste. Um teste com 100% do target sendo apenas uma das classificações. Basicamente o modelo criado tem que superá-lo. 

```python
import numpy as np

previsoes_de_base = np.ones(540)
acuracia = accuracy_score(teste_y, previsoes_de_base) * 100
print('A acuracia é de {:.2f}%'.format(acuracia))
```

```python
sns.scatterplot(x = 'expected_hours', y = 'price', data = teste_x, hue = teste_y)
```

```python
x_min = teste_x['expected_hours'].min()
x_max = teste_x['expected_hours'].max()

y_min = teste_x['price'].min()
y_max = teste_x['price'].max()
```

```python
pixels = 100 #passos dados na área do gráfico acima

eixo_x = np.arange(x_min, x_max, (x_max - x_min) / pixels)
eixo_y = np.arange(y_min, y_max, (y_max - y_min) / pixels)
```

```python
xx, yy = np.meshgrid(eixo_x, eixo_y)
pontos = np.c_[xx.ravel(), yy.ravel()]
```

```python
z = modelo.predict(pontos)
z = z.reshape(xx.shape)
```

```python
import matplotlib.pyplot as plt

plt.contourf(xx, yy, z, alpha = 0.2)
plt.scatter(teste_x.expected_hours, teste_x.price, c = teste_y, s = 5) #se for 0, pinta de uma cor, se for 1, pinta de outra
```

### Usando o estimador não linear SVC

```python
from sklearn.svm import SVC

SEED = 5
np.random.seed(SEED) #para não configurar o parâmetro de randon state no split e no SVCLinear
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25,
                                                         stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

modelo = SVC(gamma='auto')
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print('A acuracia é de {:.2f}%'.format(acuracia))
```

```python
x_min = teste_x['expected_hours'].min()
x_max = teste_x['expected_hours'].max()
y_min = teste_x['price'].min()
y_max = teste_x['price'].max()

pixels = 100 #passos dados na área do gráfico acima
eixo_x = np.arange(x_min, x_max, (x_max - x_min) / pixels)
eixo_y = np.arange(y_min, y_max, (y_max - y_min) / pixels)

xx, yy = np.meshgrid(eixo_x, eixo_y)
pontos = np.c_[xx.ravel(), yy.ravel()]

z = modelo.predict(pontos)
z = z.reshape(xx.shape)

import matplotlib.pyplot as plt

plt.contourf(xx, yy, z, alpha = 0.2)
plt.scatter(teste_x.expected_hours, teste_x.price, c = teste_y, s = 5) #se for 0, pinta de uma cor, se for 1, pinta de outra
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

SEED = 5
np.random.seed(SEED) #para não configurar o parâmetro de randon state no split e no SVCLinear
raw_treino_x, raw_teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25,
                                                         stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

scaler = StandardScaler()
scaler.fit(treino_x)
treino_x = scaler.transform(raw_treino_x)
teste_x = scaler.transform(raw_teste_x)

modelo = SVC(gamma='auto')
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print('A acuracia é de {:.2f}%'.format(acuracia))
```
