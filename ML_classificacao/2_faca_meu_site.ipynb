```python
import pandas as pd
```

```python
uri = 'https://gist.githubusercontent.com/guilhermesilveira/1b7d5475863c15f484ac495bd70975cf/raw/16aff7a0aee67e7c100a2a48b676a2d2d142f646/projects.csv'
df = pd.read_csv(uri)
df.head()
```

```python
troca = {0:1, 1:0}
df['finished'] = df['unfinished'].map(troca)

df.head()
```

#### Criando o modelo de ML

```python
#Features
x = df[['expected_hours', 'price']]
#Target
y = df['finished']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
SEED = 20
```

```python
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state = SEED, stratify = y,  test_size = 0.25)

print('Treino com {} elementos \nTeste com {} elementos'.format(len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)

previsoes = modelo.predict(teste_x)
previsoes

acuracia = accuracy_score(teste_y, previsoes) * 100
print('A acuracia é de {:.2f}%'.format(acuracia))
```

#### O teste abaixo é feito com teste. Um teste com 100% do target sendo apenas uma das classificações. Basicamente o modelo criado tem que superá-lo. 

```python
import numpy as np

previsoes_de_base = np.ones(540)
acuracia = accuracy_score(teste_y, previsoes_de_base) * 100
print('A acuracia é de {:.2f}%'.format(acuracia))
```
