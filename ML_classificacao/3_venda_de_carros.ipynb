```python
import pandas as pd

uri = 'https://gist.githubusercontent.com/guilhermesilveira/4d1d4a16ccbf6ea4e0a64a38a24ec884/raw/afd05cb0c796d18f3f5a6537053ded308ba94bf7/car-prices.csv'
df = pd.read_csv (uri)
df.head()
```

```python
troca = {'no':0, 'yes':1}

df['sold'] = df.sold.map(troca)
df.head()
```

```python
from datetime import datetime
today = datetime.today().year
df['model_age'] = today - df['model_year']

df.head()
```

```python
df['km_per_year'] = df['mileage_per_year'] * 1.60934
df.head()
```

```python
df = df.drop(columns= ['Unnamed: 0', 'mileage_per_year', 'model_year'], axis = 1)
df.head()
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 10

#Features
x = df[['price', 'model_age', 'km_per_year']]
#Target
y = df['sold']
np.random.seed(SEED) #para não configurar o parâmetro de randon state no split e no SVCLinear

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, 
                                                        stratify = y,  
                                                        test_size = 0.25)

print('Treino com {} elementos \nTeste com {} elementos'.format(len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)

previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print('A acuracia é de {:.2f}%'.format(acuracia))
```

```python
from sklearn.dummy import DummyClassifier

dummy_stratified = DummyClassifier()

dummy_stratified.fit(treino_x, treino_y)

previsoes = dummy_stratified.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print('A acuracia é de {:.2f}%'.format(acuracia))
```

```python
dummy_most_frequent = DummyClassifier(strategy = 'most_frequent')

dummy_most_frequent.fit(treino_x, treino_y)

round(dummy_most_frequent.score(teste_x, teste_y) * 100) 

```
